**Day 1 - Hands-on Lab Exercise **

```python
import numpy as np
from scipy.stats import mode  # Import mode from SciPy


# Data
data = [12, 15, 14, 10, 18, 20, 22, 17, 19, 13, 16, 15, 14, 21, 23]

# Calculate mean
mean = np.mean(data)

# Calculate median
median = np.median(data)

# Calculate mode
mode_result = mode(data)  # SciPy mode returns both the mode and its count
# Access mode value directly if it's a scalar, otherwise access the first element
mode_value = mode_result.mode if np.isscalar(mode_result.mode) else mode_result.mode[0]

# Print results
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode_value}")
```

```python
# Variance and Standard Deviation
variance = np.var(data, ddof=1)  # Sample variance
std_dev = np.std(data, ddof=1)  # Sample standard deviation

# Range
data_range = max(data) - min(data)

print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
print(f"Range: {data_range}")
```

```python
import matplotlib.pyplot as plt # Import the matplotlib.pyplot module

# Histogram
plt.hist(data, bins=5, color='blue', alpha=0.7, edgecolor='black')
plt.title("Histogram of Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Box Plot
plt.boxplot(data, vert=False)
plt.title("Box Plot of Data")
plt.xlabel("Value")
plt.show()

# Line Graph
plt.plot(data, marker='o', color='purple', linestyle='-', linewidth=2, markersize=6)
plt.axhline(y=mean, color='red', linestyle='--', label=f"Mean ({mean:.1f})")
plt.axhline(y=median, color='green', linestyle='-.', label=f"Median ({median:.1f})")
plt.axhline(y=mode_value, color='blue', linestyle=':', label=f"Mode ({mode_value})")
plt.title("Line Graph of Data")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.grid(True, alpha=0.5)
plt.show()
```

**Real World Dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Read data from a CSV file
file_path = 'position_salary.csv'
data = pd.read_csv(file_path, encoding='latin-1')

# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
print(data.head())

# Step 2: Descriptive Statistics
print("\nDescriptive Statistics:")
print(data.describe())

# Measures of Central Tendency and Dispersion for a specific column
column_name = 'Salary'
mean = data[column_name].mean()
median = data[column_name].median()
mode = data[column_name].mode()[0]
std_dev = data[column_name].std()
variance = data[column_name].var()

print(f"\nStatistics for {column_name}:")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {std_dev}")
print(f"Variance: {variance}")

# Step 3: Visualizations
# Histogram
plt.figure(figsize=(8, 5))
sns.histplot(data[column_name], bins=10, kde=True, color='blue', edgecolor='black')
plt.title(f"Histogram of {column_name}")
plt.xlabel(column_name)
plt.ylabel("Frequency")
plt.show()

# Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x=data[column_name], color='lightgreen')
plt.title(f"Box Plot of {column_name}")
plt.xlabel(column_name)
plt.show()


# Correlation Heatmap
plt.figure(figsize=(10, 8))
# Select only numerical columns for correlation calculation
numerical_data = data.select_dtypes(include=np.number)
correlation = numerical_data.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

```

Descriptive Statistics Insights

1. **Mean (Average Salary):**

  The average salary value gives an idea of the typical compensation in the dataset.

2. **Median (Middle Salary):**

  The median salary highlights the middle value, unaffected by extreme outliers.

3. **Mode (Most Frequent Salary):**

  The mode identifies the salary value that appears most often, useful for identifying popular salary bands or tiers.

4. **Standard Deviation (Salary Variability):**

  A high standard deviation suggests large differences in salaries across positions, while a low value implies consistent compensation.

5. **Variance:**

  High variance confirms significant dispersion in salaries, reinforcing observations from standard deviation.

**Generate Data from Different Distributions**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Load the dataset
dataset = pd.read_csv('position_salary.csv', encoding='latin-1')

# Extracting relevant columns
binomial_column = dataset['Salary']
poisson_column = dataset['Gender'].astype(str)
normal_column = dataset['Location'].astype(str)


# Plotting the distributions
plt.figure(figsize=(15, 5))

# Binomial distribution
plt.subplot(1, 3, 1)
plt.hist(binomial_column, bins=10, color='blue', alpha=0.7, edgecolor='black')
plt.title('Real-world Binomial Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Poisson distribution
plt.subplot(1, 3, 2)
plt.hist(poisson_column, bins=10, color='green', alpha=0.7, edgecolor='black')
plt.title('Real-world Poisson Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Normal distribution
plt.subplot(1, 3, 3)
plt.hist(normal_column, bins=20, color='red', alpha=0.7, edgecolor='black')
plt.title('Real-world Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Display the plots
plt.tight_layout()
plt.show()
```

```python
# Set the number of samples and sample size
num_samples = 1000
sample_size = 30

# Create an empty list to store sample means
sample_means = []

# Generate samples and calculate their means
for _ in range(num_samples):
    # Change: Sample from the 'Salary' column instead of the entire DataFrame
    sample = np.random.choice(data[column_name], size=sample_size, replace=True)
    sample_means.append(np.mean(sample))

# Plot the distribution of sample means
plt.figure(figsize=(8, 6))
sns.histplot(sample_means, bins=30, kde=True, color='blue', edgecolor='black')
plt.title(f"Distribution of Sample Means ({column_name}) - Central Limit Theorem")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

# Add vertical lines for population mean and sample mean
# Change: Calculate population mean from the 'Salary' column
population_mean = data[column_name].mean()
sample_mean = np.mean(sample_means)
plt.axvline(population_mean, color='red', linestyle='--', label=f"Population Mean: {population_mean:.2f}")
plt.axvline(sample_mean, color='green', linestyle='-', label=f"Sample Mean: {sample_mean:.2f}")
plt.legend()

plt.show()
```
